```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data.csv")

# Show data
print(df.head())

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Create new column: Revenue
df['Revenue'] = df['Price'] * df['Quantity']

# ---------------- KPIs ----------------
total_revenue = df['Revenue'].sum()
avg_order_value = df['Revenue'].mean()

print("Total Revenue:", total_revenue)
print("Average Order Value:", avg_order_value)

# ---------------- Category Analysis ----------------
category_sales = df.groupby('Category')['Revenue'].sum()
print("\nCategory Sales:\n", category_sales)

# ---------------- City Analysis ----------------
city_sales = df.groupby('City')['Revenue'].sum()
print("\nCity Sales:\n", city_sales)

# ---------------- Top Products ----------------
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
print("\nTop Products:\n", top_products)

# ---------------- Charts ----------------

# 1. Sales by Category
category_sales.plot(kind='bar', title="Sales by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()

# 2. Sales by City
city_sales.plot(kind='bar', title="Sales by City")
plt.xlabel("City")
plt.ylabel("Revenue")
plt.show()

# 3. Sales Trend
df.groupby('Date')['Revenue'].sum().plot(marker='o', title="Sales Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```
